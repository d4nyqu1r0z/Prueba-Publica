```python

from itertools import cycle
personas_registradas = []

def grabar_persona():
  rut = input("Ingrese el RUT sin digito verificador: ")
  digito= input("Ingrese digito verificador ")
  if digito_verificador(rut)==digito:
    nombre = input("Ingrese el nombre: ")
    edad = int(input("Ingrese la edad: "))
    pais_nacimiento = input("Ingrese el país de nacimiento: ")
    ciudad_nacimiento = input("Ingrese la ciudad de nacimiento: ")
    persona = [rut, nombre, edad, pais_nacimiento, ciudad_nacimiento]
    personas_registradas.append(persona)
    print("Se ha registrado correctamente.")
  else:
    print("Datos incorrectos")

def digito_verificador(rut):
  reversed_digits = map(int, reversed(str(rut)))
  factors = cycle(range(2, 8))
  s = sum(d * f for d, f in zip(reversed_digits, factors))
  return (-s) % 11

def buscar_persona():
  rut = input("Ingrese el RUT de la persona que desea buscar: ")
  for persona in personas_registradas:
    if persona[0] == rut:
      print("Informacion:")
      print("Rut:", persona[0])
      print("Nombre: ", persona[1])
      print("Edad: ", persona[2])
      print("Pais de nacimiento: ", persona[3])
      print("Ciudad de nacimiento: ", persona[4])

def imprimir_certificado():
  certificados = ["Certificado de nacimiento", "Certificado de estado conyugal"]
  for persona in personas_registradas:
    print("Certificado de", persona[1], "con RUT", persona[0])
    for certificado in certificados:
      print("Certificado: ", certificado)
      print("RUT: ", persona[0])
      print("Nombre: ", persona[1])

def eliminar_persona():
  rut = input("Ingresa el rut de la persona que deseas eliminar: ")
  for persona in personas_registradas:
    if persona[0] == rut:
      personas_registradas.remove(persona)
      print("Ha sido eliminado exitosamente")
      return
    else:
      print("Esa persona no se encuentra registrada")

def salir():
  print("Gracias por usar el programa. ¡Hasta luego!")

def menu():
  while True:
        print("----- MENÚ -----")
        print("1. Grabar")
        print("2. Buscar")
        print("3. Imprimir certificados")
        print("4. Eliminar")
        print("5. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
          grabar_persona()
        elif opcion == "2":
          buscar_persona()
        elif opcion == "3":
          imprimir_certificado()
        elif opcion == "4":
          eliminar_persona()
        elif opcion == "5":
          salir()
          break
        else:
            print("Opción inválida. Intente nuevamente.")

menu()
```

```python
from itertools import cycle
def digito_verificador(dni):
    dni = str(dni)
    reversed_digits = map(int, reversed(str(dni)))
    factors = cycle(range(2, 8))
    s = sum(d * f for d, f in zip(reversed_digits, factors))
    if (-s) % 11 == 10:
        return "K"
    return (-s) % 11

def validar_dni(dni):
    dni = str(dni) + "-" + str(digito_verificador(dni))
    return dni

opciones = ["Guardar registro", "Buscar Registro", "Imprimir Certificado", "Eliminar Registro", "Salir"]
usuarios_registrados = {}
Certificados = {
    "Nacimiento" : "El Registro Civil certifica que el usuario {nombre} {apellido} \nidentificado con dni {dni} nacio en {pais} en {ciudad} y tiene {edad} años",
    "Matrimonio" : "El Registro Civil certifica que el usuario {nombre} {apellido} \nidentificado con dni {dni} esta casado con {conyuge}",
}
opcion = 0

print("\n\t\t----------BIENVENIDO AL TOTEM DE REGISTRO CIVIL----------\n")

while opcion != 5:
    print("Por favor seleccione la opcion que desea realizar: \n")
    for i in range(len(opciones)):
        print(f"{i + 1}) {opciones[i]}")
    opcion = int(input("Seleccione la opcion que desea realizar: "))

    if opcion == 1:
        print("\n\t\t----------GUARDAR REGISTRO----------\n")
        dni = int(input("Ingrese su dni sin puntos ni guion y sin el ultimo digito : "))
        nombre = input("Ingrese su nombre: ").capitalize()
        apellido = input("Ingrese su apellido: ").capitalize()
        edad = int(input("Ingrese su edad: "))
        pais = input("Ingrese su pais de origen: ").capitalize()
        ciudad = input("Ingrese su ciudad de nacimiento: ").capitalize()
        if edad >= 18:
            estado_civil = input("Ingrese su estado civil: ").capitalize()
        else:
            estado_civil = None

        dni = validar_dni(dni)
        usuarios_registrados[dni] = [nombre, apellido, edad, pais,ciudad,estado_civil]
        print(usuarios_registrados)
        print("\n\t\t----------REGISTRO GUARDADO CON EXITO----------\n")
        continue

    elif opcion == 2:
        print("\n\t\t----------BUSCAR REGISTRO----------\n")
        dni = int(input("Ingrese su dni sin puntos ni guion y sin el ultimo digito para verificar su registro: "))

        dni = validar_dni(dni)
        if dni in usuarios_registrados:
            print(f"El dni {dni} pertenece a: \n {usuarios_registrados[dni][0]} {usuarios_registrados[dni][1]} \n Edad: {usuarios_registrados[dni][2]} \n Pais: {usuarios_registrados[dni][3]} \n Ciudad: {usuarios_registrados[dni][4]} \n Estado Civil: {usuarios_registrados[dni][5]}")
        else:
            print("El dni ingresado no se encuentra registrado")

    elif opcion == 3:
        print("\n\t\t----------IMPRIMIR CERTIFICADO----------\n")
        dni = int(input("Ingrese su dni sin puntos ni guion y sin el ultimo digito para solicitar el certificado: "))

        dni = validar_dni(dni)
        if dni in usuarios_registrados:
            print("Que tipo de certificado desea solicitar: ")
            for i in range(len(Certificados)):
                print(f"{i + 1}) {list(Certificados.keys())[i]}")
            certificado = int(input("Seleccione la opción del certificado a imprimir: "))
            if certificado == 1:
                if pais == "Argentina":
                    print("\n-------------CERTIFICADO DE NACIMIENTO--------------")
                    print(Certificados["Nacimiento"].format(nombre=usuarios_registrados[dni][0], apellido=usuarios_registrados[dni][1], dni=dni, pais=usuarios_registrados[dni][3], edad=usuarios_registrados[dni][2], ciudad=usuarios_registrados[dni][4]))
                else:
                    print("El usuario no nacio en Argentina, no se puede emitir el certificado")
            elif certificado == 2:
                if estado_civil != "Casado":
                    print("El usuario no esta casado, no se puede emitir el certificado")
                    continue
                conyuge = input("Ingrese el nombre de su conyuge: ").capitalize()
                apellido_conyuge = input("Ingrese el apellido de su conyuge: ").capitalize()
                conyuge = conyuge + " " + apellido_conyuge
                print("\n\t\t----------CERTIFICADO EMITIDO CON EXITO----------\n")
                print(Certificados["Matrimonio"].format(nombre=usuarios_registrados[dni][0], apellido=usuarios_registrados[dni][1], dni=dni, conyuge=conyuge,))
        else:
            print("El dni ingresado no se encuentra registrado")

    elif opcion == 4:
        print("\n\t\t----------ELIMINAR REGISTRO----------\n")
        dni = int(input("Ingrese su dni sin puntos ni guion y sin el ultimo digito para eliminar su registro: "))

        dni = validar_dni(dni)
        if dni in usuarios_registrados:
            usuarios_registrados.pop(dni)
            print("Registro eliminado con exito")
        else:
            print("El dni ingresado no se encuentra registrado")

    else:
        print("La opcion ingresada no es valida")
        continue

print("\n\t\t----------GRACIAS POR UTILIZAR EL TOTEM REGISTRO CIVIL----------\n")
```
